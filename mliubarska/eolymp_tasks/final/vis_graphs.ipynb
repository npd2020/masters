```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
options = {'node_color': 'limegreen',
           'node_size': 1000,
           'width': 3}
```

```python
def draw_graph(root, nodes, edges):
    G = nx.Graph()
    G.add_nodes_from(nodes)
    G.add_edges_from(edges)
    nx.draw(G, with_labels=True, **options)
    plt.show()
    
    edges_bfs = nx.bfs_edges(G, root)
    nodes_bfs = [root] + [v for u, v in edges_bfs]
    print nodes_bfs
    
    G.clear()
```
